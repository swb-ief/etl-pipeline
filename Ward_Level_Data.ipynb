````python
file = "http://stopcoronavirus.mcgm.gov.in/assets/docs/Dashboard.pdf"

new_case_growth = tabula.read_pdf(file, pages=24, multiple_tables=False)
````

````python
new_case_growth[0]
````

````python
df = new_case_growth[0]
df.drop(labels=11, inplace=True)
````

### Save ward identifier with corresponding ward name in a dictionary

````python
wards = df.iloc[0][1:]
identifiers = df.iloc[0][1:].index
ward_identifiers = {}
````

````python
# for loop needs to be fixed

for ward in wards:
    text = str(ward)
    name = text.replace('\r',' ')
    index = wards[wards=='Borivali'].index[0]
    for identifier in identifiers:
        if index == identifier:
            ward_identifiers[identifier] = name
````
